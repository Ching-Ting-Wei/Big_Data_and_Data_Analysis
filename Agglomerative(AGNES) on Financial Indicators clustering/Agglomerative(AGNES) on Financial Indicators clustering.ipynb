```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
datasets = pd.read_csv('Financial Indicators_X.csv', encoding='unicode_escape')
x = datasets.drop('Company Name', axis=1)
```

```python
AGNESclustering = AgglomerativeClustering(
    affinity='euclidean', n_clusters=4, linkage='ward')
```

```python
AGNESclustering.fit(x)
labels = AGNESclustering.fit_predict(x)
```

```python
plt.figure(figsize=(20, 14))
plt.scatter(x.values[:, 0], x.values[:, 1], c=labels, cmap='rainbow')
plt.axis('equal')
plt.title('Clustering')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

linkage_matrix =linkage(x,'ward')
figure = plt.figure(figsize=(20,15))

dendrogram(linkage_matrix,labels=labels)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Instances')
plt.ylabel('Distance')
plt.tight_layout()
plt.show()
```

```python
Labels = labels

Clustering_results = pd.DataFrame(datasets)
Clustering_results["Clustering_results"] = Labels

Clustering_results.to_csv(
    'Clustering_results_AGNES.csv', mode='a', header=True)
```
